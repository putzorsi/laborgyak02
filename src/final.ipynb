```python
from bertopic import BERTopic
from bertopic.representation import KeyBERTInspired, MaximalMarginalRelevance

import os
directory = "../data/combined_transcripts_freeConv"

docs = []
meta = {}
i = 0
for filename in os.listdir(directory):
    with open(os.path.join(directory, filename), 'r', encoding="UTF-8") as f:
        txt = f.read()

    mordor = [e.strip() for e in txt.split("\n") if e.startswith("Mordor: ")]
    gondor = [e.strip() for e in txt.split("\n") if e.startswith("Gondor: ")]

    j = len(mordor)
    k = len(gondor)
    for h in range(j):
        meta[i+h] = filename + "_mordor_" + str(h)
    for g in range(k):
        meta[i+g+j] = filename + "_gondor_" + str(g)
    mordor = [e.replace("Mordor: ", "") for e in mordor]
    gondor = [e.replace("Gondor: ", "") for e in gondor]

    docs += mordor + gondor
    a = (j+k)
    i += a
print("Az adatbázis",len(docs), "beszédfordulót tartalmaz.")
print("Az adatbázis első beszédfordulójának metaadata:", meta[0])
```

```python
model_path = "../data/my_model_final_2024-05-16 17:36:03.092034"
final_model = BERTopic.load(model_path)
print("A modellt betöltöttük")
```

```python
import os
import itertools

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


in_dir = "../data/combined_transcripts_freeConv"
txts = [f for f in os.listdir(in_dir) if os.path.isfile(os.path.join(in_dir, f))]

def topic2image(txt, image_path):

    mordor = []
    gondor = []

    for k,v in meta.items():
        if txt in v:
            if "mordor" in v:
                mordor.append(topic_distr[k])
            else:
                gondor.append(topic_distr[k])
    combined = np.asarray(list(itertools.chain.from_iterable(zip(gondor, mordor)))).T
    img = Image.fromarray(combined, "L")
    #plt.imshow(img)
    #plt.show()
    img = img.resize((10000, 2500))
    img.save(image_path, "PNG")
```

```python
import os
import numpy as np

dials = "../data/combined_transcripts_freeConv"

txts = [f for f in os.listdir(dials) if os.path.isfile(os.path.join(dials, f))]
txt2tdist = []
for txt in txts:
    txt_docs = [k for k,v in meta.items() if txt in v]
    m = []
    m_vec = [0.0] * 49
    for td in txt_docs:
        tdist = topic_distr[td]
        m.append(tdist)
        m_vec = np.mean(np.asarray([m_vec, tdist]), axis=0)
    txt2tdist.append(m_vec)
    mean_vec = np.mean(np.asarray(m), axis=0)
```

```python
import pandas as pd

data_dict = {"filename": [],
             "topic": [],
             "prop": [],}
i = 0
for e in txt2tdist:
    sorted_index_array = np.argsort(e)
    print("-"*20)
    print(txts[i])
    first = sorted_index_array[-1]
    second = sorted_index_array[-2]
    third = sorted_index_array[-3]
    data_dict["topic"].append(first)
    data_dict["prop"].append(e[sorted_index_array[-1]])
    data_dict["filename"].append(txts[i])

    data_dict["topic"].append(second)
    data_dict["prop"].append(e[sorted_index_array[-2]])
    data_dict["filename"].append(txts[i])

    data_dict["topic"].append(third)
    data_dict["prop"].append(e[sorted_index_array[-3]])
    data_dict["filename"].append(txts[i])

    data_dict["topic"].append(99)
    data_dict["prop"].append(1.0 - sum([e[sorted_index_array[-1]], e[sorted_index_array[-2]], e[sorted_index_array[-3]]]))
    data_dict["filename"].append(txts[i])

    print("first topic", first, "proportion", e[sorted_index_array[-1]]*100)
    print("second topic", second, "proportion", e[sorted_index_array[-2]]*100)
    print("third topic", third, "proportion", e[sorted_index_array[-3]]*100)
    i += 1
```

```python
data_dict["CustomName"] = [topic2customname[e] for e in data_dict["topic"]]
```
